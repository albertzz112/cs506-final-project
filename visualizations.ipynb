# Visualizations

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Ensure plots are displayed within the notebook
%matplotlib inline
```

```python
# Load the dataset
df = pd.read_csv("data/data_cleaned.csv")
```

```python
df.head()
```

```python
#Distribution of Arrival Delays
plt.figure(figsize=(10, 6))
sns.histplot(df['ARR_DELAY'], bins=30, kde=True)
plt.title('Distribution of Arrival Delays')
plt.xlabel('Arrival Delay (minutes)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
#Average Arrival Delay by Airline
plt.figure(figsize=(12, 6))
avg_delay_airline = df.groupby('AIRLINE')['ARR_DELAY'].mean().sort_values()
sns.barplot(x=avg_delay_airline.values, y=avg_delay_airline.index)
plt.title('Average Arrival Delay by Airline')
plt.xlabel('Average Arrival Delay (minutes)')
plt.ylabel('Airline')
plt.grid()
plt.show()
```

```python
#Arrival Delay by Day of Week
plt.figure(figsize=(12, 6))
sns.boxplot(x='DAY_OF_WEEK', y='ARR_DELAY', data=df)
plt.title('Arrival Delay by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Arrival Delay (minutes)')
plt.grid()
plt.show()
```

```python
#Heatmap of Average Arrival Delay by Month and Airline
plt.figure(figsize=(12, 8))
avg_delay_month_airline = df.groupby(['MONTH', 'AIRLINE'])['ARR_DELAY'].mean().unstack()
sns.heatmap(avg_delay_month_airline, cmap='coolwarm', annot=True, fmt=".1f")
plt.title('Average Arrival Delay by Month and Airline')
plt.xlabel('Airline')
plt.ylabel('Month')
plt.show()
```

```python
#Scatter Plot of Distance vs. Arrival Delay
plt.figure(figsize=(10, 6))
sns.scatterplot(x='DISTANCE', y='ARR_DELAY', data=df, alpha=0.6)
plt.title('Scatter Plot of Distance vs. Arrival Delay')
plt.xlabel('Distance (miles)')
plt.ylabel('Arrival Delay (minutes)')
plt.grid()
plt.show()
```

```python
#Line Plot of Average Arrival Delay Over Months
plt.figure(figsize=(10, 6))
avg_delay_month = df.groupby('MONTH')['ARR_DELAY'].mean()
sns.lineplot(x=avg_delay_month.index, y=avg_delay_month.values, marker='o')
plt.title('Average Arrival Delay Over Months')
plt.xlabel('Month')
plt.ylabel('Average Arrival Delay (minutes)')
plt.xticks(avg_delay_month.index)
plt.grid()
plt.show()
```
